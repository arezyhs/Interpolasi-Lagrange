## ðŸ“Œ Tujuan

Tujuan dari proyek ini adalah untuk memperkirakan nilai emisi COâ‚‚ pada penggunaan listrik yang tidak terdapat dalam dataset (dalam hal ini, `X = 213`), menggunakan interpolasi Lagrange. Dengan metode ini, kita bisa:

3. Mengevaluasi tingkat akurasi hasil prediksi menggunakan metrik error seperti **MAPE (Mean Absolute Percentage Error)**, **Error Relatif**, **R-squared**, dan **Batas Error Lagrange**.

---

## Metode Interpolasi Lagrange

**Interpolasi Lagrange** membentuk persamaan polinomial yang melewati titik-titik data terdekat. Orde interpolasi menentukan jumlah titik data yang digunakan:
- **Orde 1**: Menggunakan 2 titik terdekat untuk membentuk polinomial linier.
- **Orde 2**: Menggunakan 3 titik terdekat untuk membentuk polinomial kuadrat.
- **Orde 3**: Menggunakan 4 titik terdekat untuk membentuk polinomial kubik.
- **Orde 4-6**: Menggunakan lebih banyak titik untuk membentuk polinomial dengan derajat yang lebih tinggi.

Variabel `set_order` dalam kode digunakan untuk mengatur orde interpolasi.

## ðŸ”§ Implementasi Kode

Kode diimplementasikan menggunakan Python dengan pustaka `numpy`, `sympy`, `matplotlib`, dan `pandas`. Berikut adalah deskripsi fungsi utama:

- **`lagrange_polynomial`**: Membentuk polinomial Lagrange simbolik berdasarkan titik data terdekat.
- **`lagrange_interpolation`**: Menghasilkan nilai prediksi `Y` pada titik `X` target dengan orde yang ditentukan.
- **`lagrange_error_bound_manual`**: Menghitung batas error interpolasi Lagrange secara manual.
- **`calculate_error_metrics`**: Menghitung metrik error untuk mengevaluasi hasil interpolasi.

---

## ðŸ“Š Evaluasi Hasil

Setelah interpolasi dan prediksi dilakukan, beberapa metrik evaluasi berikut digunakan untuk mengukur akurasi prediksi:

- **Error Relatif**: Rata-rata persentase kesalahan prediksi.
- **MAPE** (Mean Absolute Percentage Error): Persentase rata-rata kesalahan absolut prediksi dibandingkan nilai aktual.
- **R-Squared**: Mengukur kesesuaian hasil prediksi terhadap data aktual.
- **Batas Error Lagrange**: Batas error maksimum interpolasi Lagrange untuk setiap orde.

## ðŸ“Œ Kesimpulan

Interpolasi Lagrange sangat efektif dalam menghasilkan prediksi berdasarkan titik data terbatas. Dalam studi kasus ini:
- Orde interpolasi yang lebih tinggi umumnya memberikan hasil yang lebih baik untuk data aktual (nilai R-Squared yang lebih tinggi).
- Penggunaan orde yang lebih tinggi dapat menyebabkan **overfitting** jika diterapkan pada dataset yang terlalu kecil.

Pengguna dapat memilih orde interpolasi yang sesuai dengan kompleksitas dan kebutuhan akurasi data yang diinginkan.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import pandas as pd
import math
from decimal import Decimal, getcontext

# Set precision untuk Decimal dan pengaturan tampilan pandas
getcontext().prec = 6
pd.options.display.float_format = '{:.6f}'.format

# Data untuk Electricity (TWh) dan CO2 Emission (kt)
electricity_data = np.array([189.66, 203.11, 209.71, 217.97, 222.64, 235.41, 247.39])
co2_emission_data = np.array([447940.0024, 483910.0037, 488549.9878, 482510.0098, 517320.0073, 576989.9902, 619840.0269])

# Membuat variabel simbolik untuk x
x = sp.symbols('x')

# Fungsi untuk membentuk polinomial Lagrange simbolik berdasarkan titik terdekat
def lagrange_polynomial(x_values, y_values):
    n = len(x_values)
    polynomial = 0
    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        polynomial += term
    return sp.expand(polynomial)

# Fungsi interpolasi polinomial Lagrange
def lagrange_interpolation(x_values, y_values, x_target, order):
    idx = np.argsort(np.abs(x_values - x_target))[:order + 1]
    x_selected = x_values[idx]
    y_selected = y_values[idx]
    lagrange_poly = lagrange_polynomial(x_selected, y_selected)
    interpolated_value = sp.lambdify(x, lagrange_poly, 'numpy')(x_target)
    return interpolated_value, lagrange_poly

# Fungsi untuk menghitung batas error Lagrange
def lagrange_error_bound_manual(x_values, y_values, x_target, order):
    idx = np.argsort(np.abs(x_values - x_target))[:order + 2]
    x_selected = x_values[idx]
    y_selected = y_values[idx]

    if order + 1 < len(x_selected):
        derivative_max = 0
        for i in range(order + 2):
            for j in range(i + 1, order + 2):
                try:
                    diff_y = (y_values[j] - y_values[i])
                    diff_x = (x_values[j] - x_values[i])
                    derivative_max = max(derivative_max, abs(diff_y / diff_x))
                except ZeroDivisionError:
                    pass
    else:
        derivative_max = 0

    product_term = 1
    for xi in x_selected[:order + 1]:
        product_term *= abs(x_target - xi)

    error_estimate = abs(derivative_max * product_term / math.factorial(order + 1))
    return error_estimate

# Fungsi metrik error
def calculate_error_metrics(y_actual, y_pred):
    error_relatif = np.mean(np.abs(y_actual - y_pred) / y_actual) * 100
    R = np.corrcoef(y_actual, y_pred)[0, 1]
    R_squared = R ** 2
    MAPE = np.mean(np.abs((y_actual - y_pred) / y_actual)) * 100
    return error_relatif, R, R_squared, MAPE

# Nilai x target yang dicari
x_target = 213

# Variabel untuk memilih orde interpolasi
set_order = 6  # Ubah ini sesuai dengan orde yang diinginkan
max_order = min(6, len(electricity_data) - 1)
if set_order > max_order:
    print(f"Orde yang diminta ({set_order}) melebihi max_order ({max_order}). Menggunakan max_order.")
    set_order = max_order

# Proses interpolasi dan evaluasi error untuk orde yang ditentukan
higher_order_predictions = {}
lagrange_polynomials = {}
error_bounds = {}
results_tables = {}

for order in range(1, set_order + 1):
    prediction, poly = lagrange_interpolation(electricity_data, co2_emission_data, x_target, order)
    higher_order_predictions[order] = prediction
    lagrange_polynomials[order] = poly
    error_bounds[order] = lagrange_error_bound_manual(electricity_data, co2_emission_data, x_target, order)
    y_pred = np.array([sp.lambdify(x, lagrange_polynomials[order], 'numpy')(xi) for xi in electricity_data])
    galat = np.abs(co2_emission_data - y_pred)

    # Tabel hasil prediksi dan galat untuk setiap orde
    results_table = pd.DataFrame({
        "Electricity (TWh)": electricity_data,
        "Y Aktual": co2_emission_data,
        f"Y Prediksi": y_pred,
        f"Galat Orde": galat
    }).round(6)
    results_tables[order] = results_table
    error_relatif, R, R_squared, MAPE = calculate_error_metrics(co2_emission_data, y_pred)

    # Tampilkan tabel hasil dan evaluasi error per orde
    print(f"\n=== Orde {order} ===")
    print("\nTabel Hasil Prediksi dan Galat:")
    print(results_table)

    print(f"\nEvaluasi Error:")
    print(f"  Error Relatif (%)    : {error_relatif:.2f}%")
    print(f"  Korelasi Pearson (R) : {R:.4f}")
    print(f"  R-Squared            : {R_squared:.4f}")
    print(f"  MAPE (%)             : {MAPE:.2f}%")
    print(f"  Batas Error Lagrange : {error_bounds[order]:.6f}")

    print(f"\nPrediksi untuk X = {x_target} adalah {higher_order_predictions[order]:.6f}")
    print("Persamaan Polinomial Lagrange:")
    sp.pretty_print(lagrange_polynomials[order])
    print("=" * 50)

# Grafik Plot Hasil
plt.figure(figsize=(12, 7))
plt.scatter(electricity_data, co2_emission_data, color='orange', s=100, label='Titik Data Asli')
x_range = np.linspace(min(electricity_data), max(electricity_data), 100)
for order in range(1, set_order + 1):
    y_interp = [lagrange_interpolation(electricity_data, co2_emission_data, x, order)[0] for x in x_range]
    plt.plot(x_range, y_interp, label=f'Interpolasi Orde {order}', linewidth=2)
    plt.scatter(x_target, higher_order_predictions[order], s=100, marker='x', color='red', label=f'Prediksi Orde {order} (Y â‰ˆ {higher_order_predictions[order]:.0f})', zorder=5)

# Settingan grafik
plt.xlabel('Electricity dari Bahan Bakar Fosil (TWh)')
plt.ylabel('Emisi CO2 (kt)')
plt.title(f'Prediksi Emisi COâ‚‚ pada X = {x_target} menggunakan Interpolasi Lagrange (Orde 1 hingga {max_order})')
plt.ylim(min(co2_emission_data) - 50000, max(co2_emission_data) + 50000)
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.show()
```
