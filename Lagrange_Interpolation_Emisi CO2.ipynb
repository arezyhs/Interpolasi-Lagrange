## ðŸ“Œ Tujuan

## Metode Interpolasi Lagrange

**Interpolasi Lagrange** membentuk persamaan polinomial menggunakan sejumlah titik data terdekat untuk mendekati nilai yang diinginkan. Prediksi dapat dilakukan dengan berbagai **orde polinomial**, di antaranya:

- **Orde 1**: Menggunakan 2 titik data terdekat untuk membentuk polinomial linier.
- **Orde 2**: Menggunakan 3 titik data terdekat untuk membentuk polinomial kuadrat.
- **Orde 3**: Menggunakan 4 titik data terdekat untuk membentuk polinomial kubik.
- **Orde 4 hingga 6**: Menggunakan lebih banyak titik untuk meningkatkan kompleksitas polinomial.

Variabel `set_order` dalam kode memungkinkan pengguna mengatur orde interpolasi yang diinginkan.

---

## ðŸ”§ Implementasi Kode

Kode ini diimplementasikan dalam Python menggunakan pustaka `numpy`, `sympy`, `matplotlib`, dan `pandas`. Berikut fungsi utama dalam kode:

- **`lagrange_polynomial`**: Membentuk polinomial Lagrange berdasarkan titik data terdekat.
- **`lagrange_interpolation`**: Menghasilkan nilai prediksi `Y` di titik `X` target sesuai orde yang ditentukan.
- **`lagrange_error_bound_manual`**: Menghitung batas error interpolasi Lagrange secara manual.
- **`calculate_error_metrics`**: Menghitung berbagai metrik error, termasuk **Error Relatif**, **MAPE**, **MAE**, **RMSE**, **R-Squared**, dan **Batas Error Lagrange**.

---

## ðŸ“Š Evaluasi Hasil

Setelah interpolasi dilakukan, evaluasi dilakukan untuk berbagai orde. Berikut adalah metrik yang digunakan:

- **Error Relatif**: Rata-rata persentase kesalahan prediksi terhadap data aktual.
- **MAPE** (Mean Absolute Percentage Error): Persentase rata-rata kesalahan absolut prediksi dibandingkan nilai aktual.
- **MAE** (Mean Absolute Error): Rata-rata kesalahan absolut.
- **RMSE** (Root Mean Square Error): Mengukur error yang dikuadratkan dari hasil prediksi.
- **R-Squared**: Menilai kesesuaian prediksi dengan data aktual.
- **Batas Error Lagrange**: Batas error maksimum yang dihasilkan interpolasi untuk setiap orde.

## ðŸ“Œ Kesimpulan

Interpolasi Lagrange adalah metode yang efektif untuk menghasilkan prediksi berdasarkan data terbatas. Berdasarkan hasil ini, kita dapat melihat bahwa:

- **Orde yang lebih tinggi** memberikan hasil prediksi yang lebih dekat dengan data aktual, seperti ditunjukkan oleh nilai **R-Squared** yang tinggi.
- **Orde yang terlalu tinggi** bisa mengakibatkan **overfitting**, terutama pada dataset kecil.

Pengguna dapat menentukan orde interpolasi yang optimal sesuai kebutuhan dan kompleksitas data yang tersedia.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import pandas as pd
import math
from decimal import Decimal, getcontext

# Set precision untuk Decimal dan pengaturan tampilan pandas
getcontext().prec = 6
pd.options.display.float_format = '{:.6f}'.format

# Data untuk Electricity (TWh) dan CO2 Emission (kt)
data_x = np.array([189.66, 203.11, 209.71, 217.97, 222.64, 235.41, 247.39])
data_y = np.array([447940.0024, 483910.0037, 488549.9878, 482510.0098, 517320.0073, 576989.9902, 619840.0269])

# Buat variabel simbolik untuk x
x = sp.symbols('x')

# Fungsi untuk membentuk polinomial Lagrange simbolik berdasarkan titik terdekat
def lagrange_polynomial(x_values, y_values):
    n = len(x_values)
    polynomial = 0
    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        polynomial += term
    return sp.expand(polynomial)

# Fungsi interpolasi polinomial Lagrange
def lagrange_interpolation(x_values, y_values, x_target, order):
    idx = np.argsort(np.abs(x_values - x_target))[:order + 1]
    x_selected = x_values[idx]
    y_selected = y_values[idx]
    lagrange_poly = lagrange_polynomial(x_selected, y_selected)
    interpolated_value = float(lagrange_poly.evalf(subs={x: x_target}))
    return interpolated_value, lagrange_poly

# Fungsi untuk menghitung batas error Lagrange
def lagrange_error_bound_manual(x_values, y_values, x_target, order):
    idx = np.argsort(np.abs(x_values - x_target))[:order + 1]
    x_selected = x_values[idx]
    if order + 1 < len(x_values):
        derivative_max = 0
        for i in range(order + 2):
            for j in range(i + 1, order + 2):
                try:
                    diff_y = (y_values[j] - y_values[i])
                    diff_x = (x_values[j] - x_values[i])
                    derivative_max = max(derivative_max, abs(diff_y / diff_x))
                except ZeroDivisionError:
                    pass
    else:
        derivative_max = 0

    product_term = 1
    for xi in x_selected:
        product_term *= abs(x_target - xi)

    error_estimate = abs(derivative_max * product_term / math.factorial(order + 1))
    return error_estimate

# Fungsi metrik error termasuk evaluasi tambahan
def calculate_error_metrics(y_actual, y_pred):
    error_relatif = np.mean(np.abs(y_actual - y_pred) / y_actual) * 100
    R = np.corrcoef(y_actual, y_pred)[0, 1]
    R_squared = R ** 2
    MAPE = np.mean(np.abs((y_actual - y_pred) / y_actual)) * 100
    MAE = np.mean(np.abs(y_actual - y_pred))  # Mean Absolute Error
    RMSE = np.sqrt(np.mean((y_actual - y_pred) ** 2))  # Root Mean Square Error
    return error_relatif, R, R_squared, MAPE, MAE, RMSE

# Pilih x_target dalam rentang data
x_target = 213.0

# Variabel untuk memilih orde interpolasi
set_order = 6  # atur orde
max_order = min(set_order, len(data_x) - 1)
if set_order > max_order:
    print(f"Orde yang diminta ({set_order}) melebihi max_order ({max_order}). Menggunakan max_order.")
    set_order = max_order

# Proses interpolasi dan evaluasi error untuk orde yang ditentukan
higher_order_predictions = {}
lagrange_polynomials = {}
error_bounds = {}
results_tables = {}

for order in range(1, set_order + 1):
    prediction, poly = lagrange_interpolation(data_x, data_y, x_target, order)
    higher_order_predictions[order] = prediction
    lagrange_polynomials[order] = poly
    error_bounds[order] = lagrange_error_bound_manual(data_x, data_y, x_target, order)
    y_pred = np.array([float(poly.evalf(subs={x: xi})) for xi in data_x])
    galat = np.abs(data_y - y_pred)
    results_table = pd.DataFrame({
        "Electricity (KWh)": data_x,
        "Emisi CO2 (Kt)": data_y,
        "Y Prediksi": y_pred,
        "Galat/Error": galat
    }).round(6)
    results_tables[order] = results_table
    error_relatif, R, R_squared, MAPE, MAE, RMSE = calculate_error_metrics(data_y, y_pred)

    # Tabel hasil dan evaluasi error untuk tiap orde
    print(f"\nTabel Hasil Prediksi dan Galat Orde {order}:")
    print(results_table)

    print(f"\nEvaluasi Error:")
    print(f"  Error Relatif (%)    : {error_relatif:.2f}%")
    print(f"  Korelasi Pearson (R) : {R:.4f}")
    print(f"  R-Squared            : {R_squared:.4f}")
    print(f"  MAPE (%)             : {MAPE:.2f}%")
    print(f"  MAE                  : {MAE:.6f}")
    print(f"  RMSE                 : {RMSE:.6f}")
    print(f"  Batas Error Lagrange : {error_bounds[order]:.6f}")

    # Menampilkan hasil prediksi, batas error, dan persamaan polinomial Lagrange
    print(f"\nPrediksi untuk X = {x_target} adalah {higher_order_predictions[order]:.6f}")
    print("Persamaan Polinomial Lagrange:")
    sp.pretty_print(lagrange_polynomials[order])
    print("=" * 50)

# Grafik Plot Hasil
plt.figure(figsize=(14, 8))
plt.scatter(data_x, data_y, color='blue', s=100, label='Titik Data Asli', zorder=5)
x_range = np.linspace(min(data_x) - 10, max(data_x) + 10, 300)  # Menambah rentang sumbu X
colors = ['orange', 'red', 'green', 'purple', 'brown', 'grey']

for order in range(1, set_order + 1):
    y_interp = [float(lagrange_polynomials[order].evalf(subs={x: xi})) for xi in x_range]
    plt.plot(x_range, y_interp, color=colors[order - 1], linestyle='-', linewidth=1.5,
             label=f'Interpolasi Orde {order}')
    plt.scatter(x_target, higher_order_predictions[order], s=100, marker='D', edgecolor='black',
                color=colors[order - 1], label=f'Prediksi Orde {order} (Y â‰ˆ {higher_order_predictions[order]:.4f})')

# Pengaturan grafik dengan batas Y dan X yang lebih luas
plt.xlabel('Electricity (TWh)')
plt.ylabel('Emisi CO2 (Kt)')
plt.title(f'Prediksi Emisi CO2 pada X = {x_target} dengan Interpolasi Lagrange (Orde 1 hingga {set_order})')

# Menambah rentang sumbu Y agar lebih terlihat
plt.ylim(min(data_y) - 100000, max(data_y) + 100000)
plt.xlim(min(data_x) - 20, max(data_x) + 20)  # Menambah rentang sumbu X

# Menambahkan format angka sumbu Y dalam ribuan
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x / 1000)}k'))

plt.legend(loc='upper left', frameon=True, fontsize=10, ncol=2)  # Legend diatur ke dua kolom untuk menghemat ruang
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.show()
```
