Proyek ini menggunakan metode **Interpolasi Polinomial Lagrange** untuk memperkirakan **laju erosi pipa** berdasarkan **laju aliran massa pasir**. Metode ini memungkinkan kita menghitung nilai yang diinginkan dari data tertentu dengan membentuk polinomial yang melewati sejumlah titik data terdekat, tanpa memerlukan seluruh data yang ada.

---

## ðŸ“ˆ Deskripsi Data

Dataset yang digunakan dalam studi kasus ini berisi hubungan antara **laju aliran massa pasir (g/s)** sebagai variabel independen (`X`) dan **laju erosi pipa (mm/year)** sebagai variabel dependen (`Y`). Berikut adalah data yang digunakan:

| 0.224                         | 0.019265464               |

| 0.597                         | 0.136998853               |
| 0.693                         | 0.188111308               |
| 0.789                         | 0.239223762               |

---

## ðŸ“Œ Tujuan

Tujuan dari studi kasus ini adalah untuk memperkirakan laju erosi pada laju aliran massa pasir yang tidak tersedia di data asli (misalnya `X = 0.500`) menggunakan interpolasi Lagrange. Dengan interpolasi ini, kita dapat:
1. Membentuk **polinomial Lagrange** dengan orde yang diinginkan untuk mendekati data.
2. Memperkirakan nilai `Y` pada titik `X = 0.500`.
3. Mengevaluasi tingkat keakuratan prediksi berdasarkan berbagai metrik error, termasuk **MAPE (Mean Absolute Percentage Error)**, **Error Relatif**, **R-squared**, dan **Batas Error Lagrange**.

---

## Metode Interpolasi Lagrange

Metode **Interpolasi Lagrange** bekerja dengan membentuk polinomial berdasarkan sejumlah titik data terdekat. Orde interpolasi menentukan jumlah titik yang digunakan, yaitu:
- **Orde 1**: Menggunakan 2 titik data terdekat (polinomial linier).
- **Orde 2**: Menggunakan 3 titik data terdekat (polinomial kuadrat).
- **Orde 3**: Menggunakan 4 titik data terdekat (polinomial kubik).
- **Orde 4**: Menggunakan semua 5 titik data (polinomial derajat 4).

Parameter `set_order` digunakan untuk menentukan orde interpolasi yang diinginkan.

## ðŸ”§ Implementasi Kode

Kode diimplementasikan dengan Python menggunakan `numpy`, `sympy`, `matplotlib`, dan `pandas`. Berikut adalah deskripsi singkat dari fungsi-fungsi utama:

- **`lagrange_polynomial`**: Membentuk polinomial Lagrange simbolik berdasarkan titik data terdekat.
- **`lagrange_interpolation`**: Menghasilkan nilai prediksi `Y` di titik `X` target berdasarkan orde yang ditentukan.
- **`lagrange_error_bound_manual`**: Menghitung batas error Lagrange secara manual.
- **`calculate_error_metrics`**: Menghitung berbagai metrik error untuk mengevaluasi hasil interpolasi, termasuk MAPE, MAE, RMSE, dan lainnya.

## ðŸ“Š Evaluasi Hasil

Berikut adalah beberapa metrik evaluasi yang digunakan untuk menilai akurasi hasil interpolasi:

- **Error Relatif**: Persentase rata-rata kesalahan prediksi dibandingkan dengan data aktual.
- **MAPE** (Mean Absolute Percentage Error): Persentase rata-rata kesalahan absolut prediksi terhadap nilai aktual.
- **MAE** (Mean Absolute Error): Rata-rata kesalahan absolut.
- **RMSE** (Root Mean Square Error): Akar kuadrat dari rata-rata kesalahan kuadrat, lebih sensitif terhadap outlier.
- **R-Squared**: Koefisien determinasi, mengukur kesesuaian hasil prediksi dengan data aktual.
- **Batas Error Lagrange**: Batas maksimum error yang dihasilkan oleh interpolasi Lagrange.

## ðŸ“Œ Kesimpulan

Interpolasi Lagrange memberikan hasil prediksi yang akurat dengan menggunakan sejumlah kecil titik data. Berdasarkan studi kasus ini:
- Orde interpolasi yang lebih tinggi cenderung memberikan hasil yang lebih sesuai dengan data (nilai R-squared mendekati 1).
- Orde yang lebih tinggi bisa menyebabkan **overfitting** jika diterapkan pada dataset yang kecil, seperti dalam kasus ini.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import pandas as pd
import math
from decimal import Decimal, getcontext

# Set precision untuk Decimal dan pengaturan tampilan pandas
getcontext().prec = 6
pd.options.display.float_format = '{:.6f}'.format

# Data untuk Laju Aliran Massa Pasir (g/s) dan Laju Erosi Pipa (mm/year)
data_x = np.array([0.224, 0.420, 0.597, 0.693, 0.789])
data_y = np.array([0.019265464, 0.078132158, 0.136998853, 0.188111308, 0.239223762])

# Membuat variabel simbolik untuk x
x = sp.symbols('x')

# Fungsi untuk membentuk polinomial Lagrange simbolik berdasarkan titik terdekat
def lagrange_polynomial(x_values, y_values):
    n = len(x_values)
    polynomial = 0
    for i in range(n):
        term = y_values[i]
        for j in range(n):
            if i != j:
                term *= (x - x_values[j]) / (x_values[i] - x_values[j])
        polynomial += term
    return sp.expand(polynomial)

# Fungsi interpolasi polinomial Lagrange
def lagrange_interpolation(x_values, y_values, x_target, order):
    idx = np.argsort(np.abs(x_values - x_target))[:order + 1]
    x_selected = x_values[idx]
    y_selected = y_values[idx]
    lagrange_poly = lagrange_polynomial(x_selected, y_selected)
    interpolated_value = sp.lambdify(x, lagrange_poly, 'numpy')(x_target)
    return interpolated_value, lagrange_poly

# Fungsi untuk menghitung batas error Lagrange
def lagrange_error_bound_manual(x_values, y_values, x_target, order):
    idx = np.argsort(np.abs(x_values - x_target))[:order + 1]
    x_selected = x_values[idx]

    if order + 1 < len(x_values):
        derivative_max = 0
        for i in range(order + 2):
            for j in range(i + 1, order + 2):
                try:
                    diff_y = (y_values[j] - y_values[i])
                    diff_x = (x_values[j] - x_values[i])
                    derivative_max = max(derivative_max, abs(diff_y / diff_x))
                except ZeroDivisionError:
                    pass
    else:
        derivative_max = 0

    product_term = 1
    for xi in x_selected:
        product_term *= abs(x_target - xi)

    error_estimate = abs(derivative_max * product_term / math.factorial(order + 1))
    return error_estimate

# Fungsi metrik error termasuk evaluasi tambahan
def calculate_error_metrics(y_actual, y_pred):
    error_relatif = np.mean(np.abs(y_actual - y_pred) / y_actual) * 100
    R = np.corrcoef(y_actual, y_pred)[0, 1]
    R_squared = R ** 2
    MAPE = np.mean(np.abs((y_actual - y_pred) / y_actual)) * 100
    MAE = np.mean(np.abs(y_actual - y_pred))  # Mean Absolute Error
    RMSE = np.sqrt(np.mean((y_actual - y_pred) ** 2))  # Root Mean Square Error
    return error_relatif, R, R_squared, MAPE, MAE, RMSE

# Nilai x target yang dicari
x_target = 0.500

# Variabel untuk memilih orde interpolasi
set_order = 4  # Ubah ini sesuai dengan orde yang diinginkan
max_order = min(4, len(data_x) - 1)
if set_order > max_order:
    print(f"Orde yang diminta ({set_order}) melebihi max_order ({max_order}). Menggunakan max_order.")
    set_order = max_order

# Proses interpolasi dan evaluasi error untuk orde yang ditentukan
higher_order_predictions = {}
lagrange_polynomials = {}
error_bounds = {}
results_tables = {}

for order in range(1, set_order + 1):
    prediction, poly = lagrange_interpolation(data_x, data_y, x_target, order)
    higher_order_predictions[order] = prediction
    lagrange_polynomials[order] = poly
    error_bounds[order] = lagrange_error_bound_manual(data_x, data_y, x_target, order)
    y_pred = np.array([sp.lambdify(x, lagrange_polynomials[order], 'numpy')(xi) for xi in data_x])
    galat = np.abs(data_y - y_pred)
    results_table = pd.DataFrame({
        "Laju Aliran Massa Pasir (g/s)": data_x,
        "Y Aktual": data_y,
        "Y Prediksi": y_pred,
        "Galat/Error": galat
    }).round(6)
    results_tables[order] = results_table
    error_relatif, R, R_squared, MAPE, MAE, RMSE = calculate_error_metrics(data_y, y_pred)

    # Tabel hasil dan evaluasi error untuk tiap orde
    print(f"\n=== Orde {order} ===")
    print("\nTabel Hasil Prediksi dan Galat:")
    print(results_table)

    print(f"\nEvaluasi Error:")
    print(f"  Error Relatif (%)    : {error_relatif:.2f}%")
    print(f"  Korelasi Pearson (R) : {R:.4f}")
    print(f"  R-Squared            : {R_squared:.4f}")
    print(f"  MAPE (%)             : {MAPE:.2f}%")
    print(f"  MAE                  : {MAE:.6f}")
    print(f"  RMSE                 : {RMSE:.6f}")
    print(f"  Batas Error Lagrange : {error_bounds[order]:.6f}")

    # Menampilkan hasil prediksi, batas error, dan persamaan polinomial Lagrange
    print(f"\nPrediksi untuk X = {x_target} adalah {higher_order_predictions[order]:.6f}")
    print("Persamaan Polinomial Lagrange:")
    sp.pretty_print(lagrange_polynomials[order])
    print("=" * 50)

# Grafik Plot Hasil
plt.figure(figsize=(12, 7))
plt.scatter(data_x, data_y, color='orange', s=100, label='Titik Data Asli')
x_range = np.linspace(min(data_x), max(data_x), 100)
for order in range(1, set_order + 1):
    y_interp = [lagrange_interpolation(data_x, data_y, x, order)[0] for x in x_range]
    plt.plot(x_range, y_interp, label=f'Interpolasi Orde {order}', linewidth=2)
    plt.scatter(x_target, higher_order_predictions[order], s=100, marker='x', color='red', label=f'Prediksi Orde {order} (Y â‰ˆ {higher_order_predictions[order]:.6f})', zorder=5)

# Settingan grafik
plt.xlabel('Laju Aliran Massa Pasir (g/s)')
plt.ylabel('Laju Erosi Pipa (mm/year)')
plt.title(f'Prediksi Laju Erosi pada X = {x_target} dengan Interpolasi Lagrange')
plt.ylim(min(data_y) - 0.05, max(data_y) + 0.05)
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.show()
```
